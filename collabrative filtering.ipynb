```python
import pandas as pd
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate

# Load the dataset (ensure this CSV has the right columns)
ratings = pd.read_csv('/content/ratings_small[1].csv')  # Use a suitable ratings dataset

# Display the first few rows of the dataset
print(ratings.head())

# Create a Reader object
reader = Reader()

# Load the data into the surprise dataset
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

# Define the SVD algorithm
svd = SVD()

# Perform cross-validation
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
import pandas as pd
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate, KFold

# Load the dataset
reader = Reader()
ratings = pd.read_csv('/content/ratings_small[1].csv')

# Load the data into the surprise dataset
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

# Define the SVD algorithm
svd = SVD()

# Set up the cross-validation procedure with KFold
kf = KFold(n_splits=5)

# Perform cross-validation
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=kf, verbose=True)
```

```python
trainset = data.build_full_trainset()
svd.fit(trainset)
```
